#### What is Apoapsis and periapsis? 
- Apoapsis is the nearest and farthest point to the orbit. Whereas, the periapsis is the lowest point of the orbit
- Next feature, we are going to calculate is difference between these two for these objects.

```python
# Difference between both of them.
# https://courses.ansys.com/index.php/courses/introduction-to-orbital-elements/lessons/orbital-elements-lesson-4/topic/semi-major-axis-a/#:~:text=The%20semi%2Dmajor%20axis%20describes,major%20axis)%20of%20the%20orbit.&text=Periapsis%20%2D%20The%20point%20in%20an,closest%20to%20the%20central%20body.
space_df['AP_DIFF'] = space_df['APOAPSIS'] - space_df['PERIAPSIS']
```

###### Semimajor axis and semi-minor axis
##### Semi-minor axcis, b, is half of the shorted diameter of all ellipsde/
##### Check for references https://astronomy.swin.edu.au/cosmos/s/Semi-minor+Axis
##### Image axis : 
#### Formula: 
  - Using the logic, b2 = a^2(1-e^2)

```python
# Calculate semi-minor axis to understand geometric of space debris
# Difference space debris object may exhibit different orbital behaviors and characteristics
space_df['SEMIMINOR_AXIS'] = space_df['SEMIMAJOR_AXIS'] * np.sqrt(1 - space_df['ECCENTRICITY']**2 )
```

#### Calculate a satellite speed by using this formula:


```python
# Gravitational parameter of the central body (e.g., Earth) in km^3/s^2
# We are going to use Earth's gravitational parameter
# Since, these satellite orbit the Earth, we will use this formula
# as the gravitational parameter 
mu = 398600.4418

space_df['ORBITAL_VELOCITY'] = np.sqrt(mu / space_df['SEMIMAJOR_AXIS'])
```
